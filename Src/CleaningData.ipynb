```python
import os
import cPickle
data_path = os.path.join('..', 'Data', 'rfam_data_full.cPickle')
data  = cPickle.load( open( data_path, "rb" ) )
data['RF00083'].keys()
```

```python
#check the longest sequence. 
l1 = -1
l2 = -2
for k in data.keys():
    for seq in data[k]['sequences'].values():
        if l1< len(seq):
            l1 = len(seq)
        cleanSeq = seq.replace('.',"").replace('-','')
        if l2 <len(cleanSeq):
            l2 = len(cleanSeq)
l1,l2
```

```python
#ACGU.
#. and - are the same thing
d = {'A':0,'C':1,'G':2,'U':3,'.':4,'-':4}
df = {'RF00083':0,'RF00128':1}

vDataStructure = []
vData = []
for k in data.keys():
    for seqk in data[k]['sequences']:
        a1 = [4 for _ in range(l1)]
        a2 = [4 for _ in range(l2)]
        seq = data[k]['sequences'][seqk]
        if 'N' in seq:
            #print seqk,k
            continue
        #Keep the structure
        for i in range(len(seq)):
            a1[i] = d[seq[i]]
        vDataStructure.append( a1 + [seqk,df[k]]) 
        #wihout the structure
        j =0 
        for i in range(len(seq)):
            if not d[seq[i]] is 4: 
                a2[j] = d[seq[i]]
                j+=1
        vData.append( a2 + [seqk,df[k]]) 
        
```

```python
#with .......----- in the sequence
import pandas as pd 
cols = ['s'+str(i) for i in range(l1)]
dfstruct = pd.DataFrame(vDataStructure,columns=cols + ['infere','familly'])
cols = ['s'+str(i) for i in range(l2)]
df = pd.DataFrame(vData,columns=cols + ['infere','familly'])
dfstruct.to_csv('seq_struct_data.csv',index_label=False)
df.to_csv('seq_data.csv',index_label=False)
```
